## Como Configurar a Chave de API (GOOGLE_API_KEY) no Google Colab

É fundamental configurar corretamente sua chave de API para que o modelo Gemini possa ser utilizado. Siga este guia passo a passo:

1.  **Obtenha sua Chave de API:**
    *   Vá para [Google AI Studio](https://aistudio.google.com/app/apikey).
    *   Crie uma nova chave de API ou copie uma existente.

2.  **Acesse o Gerenciador de Segredos do Colab:**
    *   No lado esquerdo do seu notebook Colab, procure e clique no ícone de um **cadeado (🔑)**. Este é o Gerenciador de Segredos.

3.  **Adicione um Novo Segredo:**
    *   No painel que se abre, clique em **'Adicionar novo segredo'**.

4.  **Configure o Nome do Segredo:**
    *   No campo **'Nome'** (Key), digite **exatamente** `GOOGLE_API_KEY`.
    *   **Importante:** O nome deve ser escrito em letras maiúsculas, com underscores, sem aspas, e corresponder **perfeitamente** a `GOOGLE_API_KEY`.

5.  **Cole sua Chave de API:**
    *   No campo **'Valor'** (Value), cole a chave de API que você obteve do Google AI Studio.

6.  **Habilite o Acesso ao Notebook:**
    *   Certifique-se de que a opção **'Acesso ao notebook'** (Notebook access) esteja **ativada** (o interruptor deve estar azul ou na posição 'ligado').

7.  **Salve as Alterações:**
    *   O segredo será salvo automaticamente ou você precisará clicar em 'Adicionar' ou 'Fechar' dependendo da versão da interface.

8.  **Execute a Célula de Configuração Novamente:**
    *   Volte ao seu notebook e **execute a célula de código (`dcf758c0`)** que contém `GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')`.
    *   Se tudo estiver correto, você verá a mensagem: `GOOGLE_API_KEY successfully retrieved from Colab secrets.` e `Gemini API configured.`

Após seguir esses passos, você poderá prosseguir com as chamadas à API do Gemini.

Before you can make any API calls, you need to initialize the Generative Model.

```python
# Initialize the Gemini API
gemini_model = genai.GenerativeModel('gemini-1.5-flash-latest')
```

```python
try:
    # Make a simple test call to the model
    test_response = gemini_model.generate_content("Say hi.")
    print("API test successful! Model responded:")
    print(test_response.text)
except Exception as e:
    print(f"API test failed. Please check your API key configuration. Error: {e}")
```

Now, let's generate the motivational text based on your instructions.

```python
prompt = """Agente IA – Motivação e Engajamento para Equipes de Suporte
Objetivo
Gerar textos motivacionais e persuasivos para colaboradores que não atuam diretamente em vendas, demonstrando como suas atividades impactam positivamente os resultados comerciais da empresa e fortalecem o sentimento de pertencimento e engajamento.
Instruções para o Agente
Crie um texto motivacional com tom positivo, inspirador e profissional. O conteúdo deve mostrar que todas as áreas da empresa contribuem para o sucesso das vendas, mesmo sem contato direto com clientes ou negociações comerciais.
Diretrizes

Destaque a importância das áreas de suporte, atendimento, logística, administrativo e demais funções de apoio.
Explique de forma simples como cada atividade influencia a experiência do cliente e os resultados da empresa.
Utilize exemplos práticos do dia a dia.
Reforce a ideia de trabalho em equipe e interdependência entre as áreas.
Valorize o esforço e a dedicação dos colaboradores.
Mostre que cada entrega, processo e atendimento contribui para a satisfação do cliente e para o crescimento do negócio.
Mantenha um tom encorajador e inspirador.
Evite linguagem excessivamente técnica.
Evite críticas, comparações negativas ou desvalorização de qualquer área.
Limite o texto entre 150 e 300 palavras.

Estrutura Recomendada

Abertura inspiradora.
Explicação do impacto das atividades indiretas nas vendas.
Exemplos práticos de contribuição.
Mensagem de valorização da equipe.
Encerramento motivacional.

Exemplo de Saída
"Quando pensamos em vendas, é comum imaginar apenas os profissionais que estão em contato direto com os clientes. No entanto, cada resultado alcançado pela empresa é fruto de um esforço coletivo.
As equipes de atendimento, logística, administrativo e suporte desempenham um papel fundamental para que a experiência do cliente seja positiva. Um pedido entregue no prazo, um atendimento ágil, um processo bem executado ou uma informação correta fornecida no momento certo ajudam a construir confiança e fortalecem o relacionamento com os clientes.
Por trás de cada venda realizada existe uma rede de profissionais comprometidos em garantir que tudo funcione da melhor forma possível. Quando a logística assegura uma entrega eficiente, quando o administrativo mantém os processos organizados ou quando o suporte resolve uma demanda rapidamente, o cliente percebe qualidade e profissionalismo.
Por isso, o sucesso comercial não pertence apenas a uma área, mas a todos que contribuem diariamente para a excelência da empresa. Seu trabalho faz diferença, gera valor e ajuda a transformar oportunidades em resultados.
Continue desempenhando suas atividades com dedicação e orgulho. Cada ação conta e cada colaborador é parte fundamental do crescimento e das conquistas da equipe."""

response = gemini_model.generate_content(prompt)
print(response.text)
```
