### 💰 Orçamento Mensal Modelo em Python

Este modelo básico de orçamento mensal em Python permite que você visualize suas receitas e despesas. Você pode ajustar os valores conforme sua realidade.

---

#### 1. Definição de Receitas Mensais

Começamos listando todas as suas fontes de renda líquida para o mês.

```python
# Defina suas receitas mensais
receitas = {
    'Salário Líquido': 3500.00,  # Exemplo de salário após impostos
    'Renda Extra (Freelance)': 200.00,
    'Outras Receitas': 50.00
}

total_receitas = sum(receitas.values())

print("--- Receitas Mensais ---")
for fonte, valor in receitas.items():
    print(f"{fonte}: R$ {valor:.2f}")
print(f"\nTotal de Receitas: R$ {total_receitas:.2f}")
```

#### 2. Definição de Despesas Fixas

Despesas fixas são aquelas que geralmente não mudam muito de mês para mês e são essenciais para sua vida (aluguel, contas de consumo, etc.).

```python
# Defina suas despesas fixas mensais
despesas_fixas = {
    'Aluguel/Financiamento': 1200.00,
    'Condomínio': 300.00,
    'Luz': 150.00,
    'Água': 80.00,
    'Internet': 100.00,
    'Celular': 60.00,
    'Transporte (Passe Mensal)': 200.00, # Se for fixo
    'Plano de Saúde': 250.00,
    'Academia': 80.00
}

total_despesas_fixas = sum(despesas_fixas.values())

print("\n--- Despesas Fixas ---")
for despesa, valor in despesas_fixas.items():
    print(f"{despesa}: R$ {valor:.2f}")
print(f"\nTotal de Despesas Fixas: R$ {total_despesas_fixas:.2f}")
```

#### 3. Definição de Despesas Variáveis

Despesas variáveis são aquelas que flutuam mês a mês e sobre as quais você tem mais controle (alimentação, lazer, compras, etc.).

```python
# Defina suas despesas variáveis mensais (com um valor estimado para o orçamento)
despesas_variaveis = {
    'Alimentação (Mercado)': 600.00,
    'Alimentação (Restaurantes/Delivery)': 300.00,
    'Lazer/Entretenimento': 200.00,
    'Compras (Roupas, Eletrônicos, etc.)': 150.00,
    'Outras Despesas Variáveis': 100.00
}

total_despesas_variaveis = sum(despesas_variaveis.values())

print("\n--- Despesas Variáveis ---")
for despesa, valor in despesas_variaveis.items():
    print(f"{despesa}: R$ {valor:.2f}")
print(f"\nTotal de Despesas Variáveis: R$ {total_despesas_variaveis:.2f}")
```

#### 4. Resumo do Orçamento Mensal

Agora vamos calcular o total das despesas e o saldo final do seu orçamento.

```python
total_despesas = total_despesas_fixas + total_despesas_variaveis
saldo_final = total_receitas - total_despesas

print("\n--- Resumo do Orçamento Mensal ---")
print(f"Total de Receitas: R$ {total_receitas:.2f}")
print(f"Total de Despesas Fixas: R$ {total_despesas_fixas:.2f}")
print(f"Total de Despesas Variáveis: R$ {total_despesas_variaveis:.2f}")
print(f"Total Geral de Despesas: R$ {total_despesas:.2f}")

print("-----------------------------------")
if saldo_final > 0:
    print(f"🎉 Saldo Final: R$ {saldo_final:.2f} (Parabéns! Você está com saldo positivo.)")
elif saldo_final < 0:
    print(f"💸 Saldo Final: R$ {saldo_final:.2f} (Atenção! Você está com saldo negativo. Hora de revisar os gastos.)")
else:
    print(f"⚖️ Saldo Final: R$ {saldo_final:.2f} (Seu orçamento está equilibrado.)")
```

### 📉 Simulação: Redução de 10% nas Despesas Variáveis

Vamos aplicar uma redução de 10% em todas as despesas variáveis para ver como o saldo final do orçamento se comporta. Isso pode te ajudar a identificar o potencial de economia.

```python
# Simulação: Reduzir 10% de cada despesa variável
despesas_variaveis_simuladas = {k: v * 0.90 for k, v in despesas_variaveis.items()}

total_despesas_variaveis_simuladas = sum(despesas_variaveis_simuladas.values())

print("\n--- Despesas Variáveis (com -10% de Redução) ---")
for despesa, valor in despesas_variaveis_simuladas.items():
    print(f"{despesa}: R$ {valor:.2f}")
print(f"\nTotal de Despesas Variáveis Simuladas: R$ {total_despesas_variaveis_simuladas:.2f}")

total_despesas_simuladas = total_despesas_fixas + total_despesas_variaveis_simuladas
saldo_final_simulado = total_receitas - total_despesas_simuladas

print("\n--- Resumo do Orçamento Mensal (com -10% nas Despesas Variáveis) ---")
print(f"Total de Receitas: R$ {total_receitas:.2f}")
print(f"Total de Despesas Fixas: R$ {total_despesas_fixas:.2f}")
print(f"Total de Despesas Variáveis Simuladas: R$ {total_despesas_variaveis_simuladas:.2f}")
print(f"Total Geral de Despesas Simuladas: R$ {total_despesas_simuladas:.2f}")

print("-------------------------------------------------------------------")
if saldo_final_simulado > 0:
    print(f"🎉 Saldo Final Simulado: R$ {saldo_final_simulado:.2f} (Parabéns! Com a redução, você tem saldo positivo.)")
elif saldo_final_simulado < 0:
    print(f"💸 Saldo Final Simulado: R$ {saldo_final_simulado:.2f} (Atenção! Mesmo com a redução, ainda há déficit.)")
else:
    print(f"⚖️ Saldo Final Simulado: R$ {saldo_final_simulado:.2f} (Com a redução, seu orçamento está equilibrado.)")
```

### 📊 Automatizando o Registro Diário de Gastos

Para automatizar o registro, vamos usar a biblioteca `pandas`. Ela nos permite criar uma estrutura de dados (DataFrame) similar a uma planilha, onde podemos adicionar e organizar nossos gastos diários de forma eficiente.

```python
import pandas as pd
from datetime import datetime

# Vamos criar um DataFrame vazio para armazenar os gastos diários.
# Se você já tiver um arquivo CSV com gastos anteriores, pode carregá-lo aqui:
# df_gastos_diarios = pd.read_csv('meus_gastos.csv')

# Para começar do zero, criamos um DataFrame com as colunas necessárias:
df_gastos_diarios = pd.DataFrame(columns=['Data', 'Categoria', 'Valor', 'Descrição'])

print("DataFrame de gastos diários criado:")
display(df_gastos_diarios.head())
```

#### Função para Adicionar um Gasto

Esta função facilitará o registro de cada nova despesa no seu `DataFrame`.

```python
def adicionar_gasto(df, data, categoria, valor, descricao=''):
    # Garante que a data seja um objeto datetime
    if isinstance(data, str):
        data = datetime.strptime(data, '%Y-%m-%d')

    novo_gasto = pd.DataFrame([{'Data': data, 'Categoria': categoria, 'Valor': valor, 'Descrição': descricao}])
    df = pd.concat([df, novo_gasto], ignore_index=True)
    return df

print("Função 'adicionar_gasto' definida.")
```

#### Registrando Alguns Gastos de Exemplo

Vamos registrar alguns gastos para demonstrar como funciona. Você pode repetir esta célula ou chamar a função `adicionar_gasto` quantas vezes precisar para registrar suas despesas.

```python
# Adicione seus gastos aqui
df_gastos_diarios = adicionar_gasto(df_gastos_diarios, '2024-07-01', 'Alimentação (Mercado)', 150.00, 'Compras da semana')
df_gastos_diarios = adicionar_gasto(df_gastos_diarios, '2024-07-02', 'Transporte (Uber)', 25.50, 'Ida ao trabalho')
df_gastos_diarios = adicionar_gasto(df_gastos_diarios, '2024-07-03', 'Alimentação (Restaurantes/Delivery)', 45.00, 'Almoço com colegas')
df_gastos_diarios = adicionar_gasto(df_gastos_diarios, '2024-07-03', 'Lazer/Entretenimento', 60.00, 'Cinema')
df_gastos_diarios = adicionar_gasto(df_gastos_diarios, '2024-07-04', 'Alimentação (Mercado)', 80.00, 'Compras complementares')
df_gastos_diarios = adicionar_gasto(df_gastos_diarios, '2024-07-05', 'Outras Despesas Variáveis', 35.00, 'Presente')

print("Gastos adicionados:")
display(df_gastos_diarios)
```

#### Agregando Gastos Diários para o Orçamento

Agora, vamos somar esses gastos por categoria para ter um panorama das suas despesas variáveis reais, que serão usadas no cálculo final do orçamento.

```python
# Agrupar os gastos diários por categoria para obter as despesas variáveis reais
despesas_variaveis_reais = df_gastos_diarios.groupby('Categoria')['Valor'].sum().to_dict()

print("\n--- Despesas Variáveis Reais (do registro diário) ---")
for categoria, valor in despesas_variaveis_reais.items():
    print(f"{categoria}: R$ {valor:.2f}")

total_despesas_variaveis_reais = sum(despesas_variaveis_reais.values())
print(f"\nTotal de Despesas Variáveis Reais: R$ {total_despesas_variaveis_reais:.2f}")
```

#### Recalculando o Orçamento com Gastos Reais

Vamos recalcular o resumo do seu orçamento usando as despesas variáveis **reais** que você registrou, em vez das estimadas.

```python
# Recalcular o orçamento com as despesas variáveis reais
total_despesas_reais = total_despesas_fixas + total_despesas_variaveis_reais
saldo_final_recalculado = total_receitas - total_despesas_reais

print("\n--- Resumo do Orçamento Mensal (com Gastos Reais) ---")
print(f"Total de Receitas: R$ {total_receitas:.2f}")
print(f"Total de Despesas Fixas: R$ {total_despesas_fixas:.2f}")
print(f"Total de Despesas Variáveis Reais: R$ {total_despesas_variaveis_reais:.2f}")
print(f"Total Geral de Despesas Reais: R$ {total_despesas_reais:.2f}")

print("----------------------------------------------------")
if saldo_final_recalculado > 0:
    print(f"🎉 Saldo Final (Real): R$ {saldo_final_recalculado:.2f} (Parabéns! Você está com saldo positivo.)")
elif saldo_final_recalculado < 0:
    print(f"💸 Saldo Final (Real): R$ {saldo_final_recalculado:.2f} (Atenção! Você está com saldo negativo. Hora de revisar os gastos.)")
else:
    print(f"⚖️ Saldo Final (Real): R$ {saldo_final_recalculado:.2f} (Seu orçamento está equilibrado.)")
```
